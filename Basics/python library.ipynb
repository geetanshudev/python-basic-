Random Modules

```python
import random
print(random.random()) #print random number between 0.0.,0.1
print(random.randint(1,34))
print(random.uniform(1,34))
print(random.randrange(1,67))
```

Using String module

```python
import string
print(string.ascii_letters)
print(string.digits)
print(string.ascii_uppercase)
print(string.punctuation)
```

```python
sd = string.ascii_uppercase
fd = string.digits
print(sd)
print(fd)

for i in range(1,9):
    yt = random.choice(sd)
    gt = random.choice(fd)
    ty = gt+yt
    print(ty,end="")

```
